***Find Imputed Data***

```python
# Create Orignal Data and Find the value of None.
number = [5, 8, None, 7, 10, None, 6]
s = 0
count = 0
for x in number:
    if x is not None:
        s += x
        count += 1
mean_val = s/count
imputed = []
for x in number:
    if x is None:
        imputed.append(mean_val)
    else:
        imputed.append(x)
print("Mean:",mean_val)
print("Orignal Data:",number)
print("Imputed Data:",imputed)
```
