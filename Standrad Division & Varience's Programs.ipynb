***Write a program to find Standrad Division and Varience***

```python
# Find Standrad Division and Varience.
data = [48,49,50,51,52]
sum_sqrd = 0
# Find Mean_val of dataset.
mean = sum(data)/len(data)
# Length of dataset.
N = len(data)

for x in data:
    sum_sqrd += (x-mean)**2
var = sum_sqrd/N
std_div = (var)**0.5
# print Command
print("Standard Division:",std_div)
print("Varience:",var)
```
