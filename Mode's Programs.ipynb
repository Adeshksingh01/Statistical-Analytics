***Find Mode without using any library***

```python
# Create data and find Mode_value.
Data = [81,44,90,50,53,32,51,82,31,90,59,67,31]
freq = {}
for x in Data:
    freq[x] = freq.get(x,0) + 1
    max_freq = max(freq.values())
    modes = []
    for k, v in freq.items():
        if v == max_freq:
            modes.append(k)
print("Mode_Val:",modes)
```

***Find Mode using library***

```python
# Create data and find Mode_value.
Data = [81,44,90,50,53,32,51,82,31,90,59,67,31]
# Import library
import statistics
# Find Mode
mode_val = statistics.mode(Data)
# Print Command
print("Mode_Val:",mode_val)
```
