***Find Mean without using any Library***

```python
# Create data and find mean value.
Data = [81,44,90,50,53,32,51,82,31,59,67,31]
mean_val = sum(Data)/len(Data)
print("Mean_Val:",mean_val)
```

```python
# Find Mean.
Score = [15,12,45,98,56,14,52,63]
mean_val = sum(Score)/len(Score)
print("Mean_Val:",mean_val)
```

```python
# Find Mean.
Number = [12,45,96,85,49,78,65,46,56]
mean_val = sum(Number)/len(Number)
print("Mean_Val:",mean_val)
```

***Find Mean using Library***

```python
# Create data and find mean value.
Data = [81,44,90,50,53,32,51,82,31,59,67,31]
# Import library
import statistics
# Find mean
mean_val = statistics.mean(Data)
# Print Command
print("Mean_Val:",mean_val)
```

```python
Score = [15,12,45,98,56,14,52,63]
# Import library
import statistics
# Find Mean.
mean_val = statistics.mean(Score)
# Print Command
print("Mean_Val:",mean_val)
```
