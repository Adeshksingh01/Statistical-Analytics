***Find Qurtile of given Datasets***

```python
# Write a program to find Quartiles of given dataset.
data = [20,22,23,25,30,32,36]
# Arrange data in order.
data.sort()
# find length of dataset.
N = len(data)
# Median of half (Q2) datset.
if N % 2 == 0:
    Q2 = (data[N//2] + data[N//2-1]) / 2
else:
    Q2 = data[N//2]
# Lower & Upper half partitioning of dataset.
if N % 2 == 0:
    lower_half = data[:N//2]
    upper_half = data[N//2:]
else:
    lower_half = data[:N//2]
    upper_half = data[N//2+1:]
# Median of lower_half (Q1) dataset.
n1 = len(lower_half)
if n1 % 2 == 0:
    Q1 = (lower_half[n1//2] + lower_half[n1//2-1]) / 2
else:
    Q1 = lower_half[n1//2]
# Median of Upper_half (Q2) dataset.
n2 = len(upper_half)
if n2 % 2 == 0:
    Q3 = (upper_half[n2//2] + upper_half[n2//2-1]) / 2
else:
    Q3 = upper_half[n2//2]
# Print Command.
print("Qurtile of Data is:")
print("*Lower_half* (Q1):",Q1)
print("*Half* (Q2):",Q2)
print("*Upper_half* (Q3):",Q3)
```
