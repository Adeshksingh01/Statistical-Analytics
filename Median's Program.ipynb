***Find Median_Val without using any Library***

```python
# Create data and find median value.
Data = [81,44,90,50,53,32,51,82,31,59,67,31]
Data.sort()
N = len(Data)
if N % 2 == 0:
    median = (Data[N//2] + Data[N//2-1]) / 2
else:
    median = Data[N//2]
print("Median_val:",median)
```

***Find Median_Val using Library***

```python
# Create data and find median value.
Data = [81,44,90,50,53,32,51,82,31,59,67,31]
# Import library
import statistics
# Find Median_val
median_val = statistics.median(Data)
# Print Command
print("Median_Val:",median_val)
```
